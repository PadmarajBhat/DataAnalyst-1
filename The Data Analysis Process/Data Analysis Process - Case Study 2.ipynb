### Clean The Data

##### 1. Drop extraneous data

```python
df_08.drop(['Stnd', 'Underhood ID', 'FE Calc Appr', 'Unadj Cmb MPG'],axis=1, inplace=True)
```

```python
df_18.drop(['Stnd', 'Stnd Description', 'Underhood ID', 'Comb CO2'],axis=1, inplace=True)
```

##### 2. Rename Column

```python
df_08.rename(columns={'Sales Area':'Cert Region'}, inplace=True)
df_08
```

```python
for c in df_08.columns:
    df_08.rename(columns={c:c.lower().replace(' ','_')}, inplace=True)
    
df_08
```

```python
for c in df_18.columns:
    df_18.rename(columns={c:c.lower().replace(' ','_')}, inplace=True)
    
df_18
```

##### Filter only CA

```python
df_08_new = df_08[df_08.cert_region == 'CA'].drop('cert_region',axis=1)
df_18_new = df_18[df_18.cert_region == 'CA'].drop('cert_region',axis=1)
```

```python
print(df_08_new.shape, df_18_new.shape)
df_08_new.dropna(inplace=True)
df_18_new.dropna(inplace=True)
print(df_08_new.shape, df_18_new.shape)
```
