##### Q2: How much have vehicle classes improved in fuel economy (increased in mpg)?

```python
df_08_new.head()
```

```python
df_08_new.veh_class.value_counts()
```

```python
df_18_new.veh_class.value_counts()
```

```python
mpg_08 = df_08_new.groupby('veh_class').cmb_mpg.mean()
mpg_08
```

```python
mpg_18 = df_18_new.groupby('veh_class').cmb_mpg.mean()
mpg_18
```

```python
inc = (mpg_18 - mpg_08)*100 / mpg_08
inc.dropna(inplace=True)
plt.figure(figsize=(8,4))
plt.bar(inc.index, inc);
plt.xlabel("Vehicle Class")
plt.ylabel("% increment in the mpg")
plt.title("increment in the mpg b/n 2008 and 2018")
```

##### Q3: What are the characteristics of SmartWay vehicles? Have they changed over time? (mpg, greenhouse gas)

```python
df_08_new.smartway.unique(), df_18_new.smartway.unique()
#df_08_new.loc[df_08_new.veh_class == 'SmartWay',:]
```

```python
smart_08 = df_08_new.loc[df_08_new.smartway.isin(['yes', 'Yes', 'Elite'])]
print(smart_08.cmb_mpg.mean(), smart_08.greenhouse_gas_score.mean())

smart_18 = df_18_new.loc[df_18_new.smartway.isin(['yes', 'Yes', 'Elite'])]
print(smart_18.cmb_mpg.mean(), smart_18.greenhouse_gas_score.mean())

plt.bar(['2008','2018'], [smart_08.cmb_mpg.mean(), smart_18.cmb_mpg.mean()]);
plt.xlabel('Year');
plt.ylabel('Mean cmb_mpg');
plt.title('Changes in the mean cmb between 2008 and 2018');
plt.show()

plt.bar(['2008','2018'], [smart_08.greenhouse_gas_score.mean(), smart_18.greenhouse_gas_score.mean()]);
plt.xlabel('Year');
plt.ylabel('Mean greenhouse_gas_score');
plt.title('Changes in the mean greenhouse_gas_score between 2008 and 2018');
```
